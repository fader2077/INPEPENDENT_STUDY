```python
import os
data_dir = './train/Apple/'
all_img = sorted(os.listdir(data_dir))
print(len(all_img))
print(all_img[:5])
print(all_img[-5:])
```

```python
import glob
import os
images = glob.glob('./train/Apple/*')
print(images)
```

```python
finished = 0
n = 1          # 設定名稱從 1 開始
for i in images:
    os.rename(i, f'./train/Apple/apple.{n}.jpg')   # 改名時，使用字串格式化的方式進行三位數補零
    n = n + 1    # 每次重複時將 n 增加 
finished = True

if finished:
  print("done")
```
